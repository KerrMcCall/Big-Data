```python
## Exercise 1

def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

# Ask user to enter a number
user_input = int(input("Enter a number: "))

# Check if the number is prime and print the result
if is_prime(user_input):
    print("This is a prime number")
else:
    print("This is not a prime number")
```

```python
## Exercise 2

# Function to generate Fibonacci sequence
def fibonacci_sequence(n):
    sequence = [0, 1]
    while len(sequence) < n:
        sequence.append(sequence[-1] + sequence[-2])
    return sequence[:n]

# Ask the user to input the number of Fibonacci numbers to be calculated
num = int(input("Enter the number of Fibonacci numbers to be calculated and displayed: "))

# Generate and display the Fibonacci sequence
fib_sequence = fibonacci_sequence(num)
print(f"Fibonacci sequence of {num} numbers:")
for num in fib_sequence:
    print(num)
```

```python
## Exercise 3

# Function to print a decreasing triangle pattern
def print_triangle(rows):
    for i in range(rows, 0, -1):
        for j in range(i):
            print("*", end=" ")
        print()

# Specify the number of rows for the triangle
rows = int(5)

# Print the triangle pattern
print_triangle(rows)
```

```python
## Exercise 4

# Function to reverse a string and print it
def reverse_string():
    # Ask the user to input a string
    user_string = input("Enter a string: ")
    
    # Reverse the string
    reversed_string = user_string[::-1]
    
    # Print the reversed string
    print(f"Reversed string: {reversed_string}")

# Call the function
reverse_string()
```

```python
## Exercise 5

# Function to find the minimum and maximum values in an iterable
def find_min_max(numbers):
    min_value = min(numbers)
    max_value = max(numbers)
    return min_value, max_value

# Example usage
numbers = [10, 2, 33, 44, 5, 99, 1]
min_val, max_val = find_min_max(numbers)

# Print the results with additional string literals
print(f"Min value is: {min_val}")
print(f"Max value is: {max_val}")
```

```python
## Exercise 6

import random

# Function to play the 'Guess the Number' game
def guess_the_number():
    # Randomly choose a secret number between 1 and 10
    secret_number = random.randint(1, 10)
    
    print("Welcome to the 'Guess the Number' game!")
    print("I have chosen a secret number between 1 and 10.")
    
    while True:
        # Ask the user to enter their guess
        guess = int(input("Enter your guess: "))
        
        # Compare the guess with the secret number and provide hints
        if guess > secret_number:
            print("Too high! Try again.")
        elif guess < secret_number:
            print("Too low! Try again.")
        else:
            print("Congratulations! You guessed the correct number!")
            break

# Call the function to start the game
guess_the_number()
```

```python
##Exercise 8

import math

# Function to solve the series
def solve_series(x, N):
    total = 0
    for i in range(1, N + 1):
        term = x * (5 ** (2 * i)) / (i + math.factorial(i + 1))
        total += term
    return total

# Ask the user to input values of x and N
x = float(input("Enter the value of x: "))
N = int(input("Enter the value of N: "))

# Calculate and display the result of the series
result = solve_series(x, N)
print(f"The result of the series for x={x} and N={N} is: {result}")
```

```python
## Exercise 9

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Function to load an image and generate mirror reflections
def mirror_image(file_path):
    # Load the image
    image = cv2.imread(file_path)
    
    # Check if the image is successfully loaded
    if image is None:
        print(f"Error: Unable to load image from {file_path}")
        return
    
    # Convert the image from BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Generate horizontal mirror reflection
    horizontal_mirror = cv2.flip(image_rgb, 1)
    
    # Generate vertical mirror reflection
    vertical_mirror = cv2.flip(image_rgb, 0)
    
    # Display the original and mirrored images
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    plt.imshow(image_rgb)
    plt.title("Original Image")
    
    plt.subplot(1, 3, 2)
    plt.imshow(horizontal_mirror)
    plt.title("Horizontal Mirror")

    plt.subplot(1, 3, 3)
    plt.imshow(vertical_mirror)
    plt.title("Vertical Mirror")

    plt.show()

# Provide the correct path to your image file
file_path = 'C:\\Users\\up2162940\\Documents\\goblin.JPEG'

# Check if the file exists and call the function
if os.path.isfile(file_path):
    mirror_image(file_path)
else:
     print(f"Error: File not found at {file_path}")
```

```python
## Exercise 10

import numpy as np

# 1. Create a NumPy array with the given temperature data
temperature_data = np.array([22, 24, 19, 21, 23, 25, 26, 27, 24, 23, 20, 22, 21, 19, 18, 22, 23, 24, 26,
                             28, 30, 27, 25, 26, 24, 22, 21, 20, 19, 23])

# 2. Find the minimum, maximum, and average temperature
min_temp = np.min(temperature_data)
max_temp = np.max(temperature_data)
avg_temp = np.mean(temperature_data)

# 3. Convert the temperatures from Celsius to Fahrenheit and store it in a new array
temperatures_fahrenheit = (temperature_data * 9/5) + 32

# 4. Find all days where the temperature was above 25°C
days_above_25 = np.where(temperature_data > 25)[0]

# 5. Count how many days had temperatures below 20°C
days_below_20 = np.sum(temperature_data < 20)

# 6. Sort the temperature data in ascending order
sorted_temperatures = np.sort(temperature_data)

# 7. Reshape the 1D array into a 5x6 2D array representing weeks (5 weeks of 6 days each)
reshaped_temperatures = temperature_data[:30].reshape(5, 6)

# Output the results
print(f"Minimum temperature: {min_temp}°C")
print(f"Maximum temperature: {max_temp}°C")
print(f"Average temperature: {avg_temp:.2f}°C")
print(f"Temperatures in Fahrenheit: {temperatures_fahrenheit}")
print(f"Days with temperatures above 25°C: {days_above_25}")
print(f"Number of days with temperatures below 20°C: {days_below_20}")
print("Sorted temperatures in ascending order:", sorted_temperatures)
print("Reshaped 5x6 array representing weeks:")
print(reshaped_temperatures)
```
